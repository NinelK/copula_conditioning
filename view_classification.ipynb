```python
# check if all simulations are finished

sessions_classified = np.unique(a['session_id'])
mask_seen = np.zeros(len(sessions_classified))


path = 'unconditional_uniform'
postfix = '_50ms_u'

# path = 'unconditional_copula'
# postfix = '_50ms_copula'
cohorts = glob.glob(f"{path}/*")
print(cohorts)
for cohort in cohorts:
#     os.system(f"mkdir {cohort}/discard")
    for session_folder in glob.glob(cohort+'/*.pkl'):
        session = session_folder.split(f"{postfix}.")[0].split('/')[-1]
#         os.system(f"mv {session_folder} {cohort}/{session}{postfix}.pkl")
#         print(session_folder,f"{cohort}/{session}{postfix}.pkl")
        if session not in sessions_classified:
            pass
#             os.system(f"mv {session_folder} {cohort}/discard/{session}{postfix}.pkl")
        else:
            mask_seen[sessions_classified==session] = 1
print(mask_seen)
```

```python
s_thetas = [None for _ in range(len(a))]
a_thetas = [None for _ in range(len(a))]

path = 'unconditional_copula'
postfix = '_50ms_copula'
cohorts = glob.glob(f"{path}/*")
print(cohorts)
for cohort in cohorts:
    for session_folder in glob.glob(cohort+'/*.pkl'):
        session = session_folder.split(f"{postfix}.")[0].split('/')[-1]
        assert session in sessions_classified
        data = pd.read_pickle(session_folder)
        clusters_processed_wth_copulas = data['cluster_id'].to_numpy()
        clusters_classified_by_harry = a[a['session_id']==session]['cluster_id'].to_numpy()
        for cluster in clusters_classified_by_harry:
            assert cluster in clusters_processed_wth_copulas
            data_loc = (data['session_id']==session) & (data['cluster_id']==cluster)
            where = np.where(((a["session_id"]==session) & (a["cluster_id"]==cluster)).to_numpy())[0]
            assert len(where)==1
            a_loc = where[0]
            s_thetas[a_loc] = data.loc[data_loc]['theta_vel'].values[0]
            a_thetas[a_loc] = data.loc[data_loc]['theta_acc'].values[0]
```

```python
a['S|P_thetas'] = s_thetas
a['A|P_thetas'] = a_thetas
a['S|P_mean_R2'] = [None if np.all(theta==None) else (np.array(theta)**2).mean() for theta in s_thetas]
a['A|P_mean_R2'] = [None if np.all(theta==None) else (np.array(theta)**2).mean() for theta in a_thetas]
```
